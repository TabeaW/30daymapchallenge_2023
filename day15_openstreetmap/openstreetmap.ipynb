````julia
using GeoJSON
````

````julia
roads=GeoJSON.read("roads.geojson")
````

````julia
using GLMakie
````

````julia
using DataFrames
````

````julia
df=DataFrame(roads)
````

````julia
df.geometry
````

````julia
roads_to_plot=GeoJSON.LineString{2,Float32}[]
````

````julia
for l in df.geometry
    if typeof(l)==GeoJSON.LineString{2,Float32}
        push!(roads_to_plot,l)
    end
end
````

````julia
roads_to_plot
````

````julia
using GeoMakie
````

````julia
lines(GeoMakie.geo2basic.(roads_to_plot))
````

````julia
using CSV
````

````julia
roads_weather=CSV.File("swsmos_LATEST_opendata.csv")[2:end]|>DataFrame
````

````julia
ts=Union{Float64,Missing}[]
for station_forecast in eachrow(roads_weather)
    if station_forecast.TS=="---"
        push!(ts, missing)
    else
        push!(ts,parse(Float64,station_forecast.TS))
    end
end
````

````julia
roads_weather.TS_new=ts
````

````julia
dropmissing!(roads_weather)
````

````julia
using Statistics
````

````julia
roads_weather_combi=combine(groupby(roads_weather,:ID),:TS_new=>mean,:Lat,:Lon)
````

````julia
using ArchGDAL
layer=ArchGDAL.getlayer(ArchGDAL.read("../nuts5000_12-31.gk3.shape/nuts5000/NUTS5000_N1.shp"),0)
bounds=DataFrame(layer);
ArchGDAL.createcoordtrans(ArchGDAL.importURL("https://sg.geodatenzentrum.de/web_public/gdz/dokumentation/crs/gk3.prj"), ArchGDAL.importPROJ4("+proj=longlat +datum=WGS84 +no_defs +type=crs")) do transform
    for x in eachrow(bounds)
    ArchGDAL.transform!(x."", transform)
    end
end
````

````julia
fig =Figure()
Label(fig[1,1],text="Mean road surface temperature in the next 167 hours\n based on SWSMOS with motorways",fontsize=30)
ax=GeoAxis(fig[2,1],latlims=(48.5,51),lonlims=(6.5,12))
poly!(ax,GeoMakie.to_multipoly.(GeoMakie.geo2basic.(bounds."")),color=:transparent,strokecolor=:black,strokewidth=1)
s=scatter!(ax, roads_weather.Lon, roads_weather.Lat,color=roads_weather_combi.TS_new_mean,colormap=:berlin,colorrange=(3,9),highclip=cgrad(:berlin)[1.0])
lines!(ax,GeoMakie.geo2basic.(roads_to_plot),color=:green,strokewidth=2)
hidespines!(ax)
hidedecorations!(ax)
Colorbar(fig[2,2],s,label="Surface temperature in Â°C")
Label(fig[3,1],text="Data: Â© Deutscher Wetterdienst\nGeodata: Â© GeoBasis-DE / BKG (2023) and Â© OpenStreetMap and its contributors")
fig
````
